```python
mn = pd.Series(['andrew  ','bo;bo','   claire   '])
```

```python
mn
```

```python
mn.str.replace(';','').str.strip().str.capitalize()
```

```python
def cleanup(name):
    name = name.str.replace(';','')
    name = name.str.strip()
    name = name.str.capitalize()
    return name
```

```python
cleanup(mn)
```

```python
mn
```

```python
mn.apply(cleanup)
```
