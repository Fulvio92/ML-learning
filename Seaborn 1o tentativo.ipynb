```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(10,8), dpi=200)
sns.violinplot(data=df, x='reading score', y='parental level of education', bw=0.01)
plt.legend(bbox_to_anchor=(1.05,0.5))
```

```python
sns.swarmplot(data=df, x='math score', y='gender' ,size=2, hue='test preparation course', dodge=True)
plt.legend(bbox_to_anchor=(1.35,0.5))
```

```python
sns.boxenplot(data=df, x='math score', y='test preparation course', hue='gender')
```

```python
sns.jointplot(data=df, x='math score', y='reading score', hue='gender')
```

```python
sns.pairplot(data=df, hue='gender', diag_kind='hist')
```

```python
sns.catplot(data=df, x='gender', y='math score', kind='violin', col='lunch', row='race/ethnicity')
```

```python
g=sns.PairGrid(data=df, hue='gender')
g = g.map_upper(sns.scatterplot)
g = g.map_diag(sns.kdeplot)
g = g.map_lower(sns.kdeplot)
g = g.add_legend()
```
