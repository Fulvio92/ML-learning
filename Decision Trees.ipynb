```python
X = pd.get_dummies(df.drop('species', axis=1), drop_first=True)
#THey are too many columns, I don't need the species in this
```

```python
y = df['species']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
base_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
```

```python
print(classification_report(y_test, base_pred))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, base_pred)
plt.show()
```

```python
model.feature_importances_
#This are in the same order as the X features created before
```

```python
pd.DataFrame(index=X.columns, data=model.feature_importances_, columns=['Feature Importance']).sort_values('Feature Importance')
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(14,8), dpi=300)
plot_tree(model, feature_names=X.columns, filled=True)
plt.show()
```

```python
#X[2] is "flipper lenght" which was also the most important feature according to the model
```

```python
len(X_train)
#All the training samples go in the first (root) node (105+45+83 are the three species adding up to 233)
```

```python
def report_model(model):
    model_preds = model.predict(X_test)
    print(classification_report(y_test, model_preds))
    print('\n')
    plt.figure(figsize=(14,8), dpi=300)
    plot_tree(model, feature_names=X.columns, filled=True)
    plt.show()
```

```python
report_model(model)
```

```python
#Now we try different hyperparameters
```

```python
pruned_tree = DecisionTreeClassifier(max_depth=2)
```

```python
pruned_tree.fit(X_train, y_train)
```

```python
report_model(pruned_tree)
```

```python
max_leaf_tree = DecisionTreeClassifier(max_leaf_nodes=3)
```

```python
report_model(max_leaf_tree)
```
