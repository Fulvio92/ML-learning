```python
def create_ages(mu=50, sigma=13, num_samples=100, seed=42):
    np.random.seed(seed)
    #normal distribution of ages centered in 50 with deviation 13; 100 val
    sample_ages=np.random.normal(loc=mu, scale=sigma, size=num_samples)
    sample_ages=np.round(sample_ages,decimals=0)
    
    return sample_ages
    
```

```python
def percentage_missing (df):
    percent_nan = 100 * df.isnull().sum() / len(df)
    percent_nan = percent_nan[percent_nan > 0].sort_values()
    
    return percent_nan

#this function tells me in ascending order the percentage of missing data for each feature that has
#missing data
```

```python
plt.figure(figsize=(8,12), dpi=200)
sns.boxplot(data=df, x='Lot Frontage', y='Neighborhood', orient = 'h')
plt.show()
```

```python
#It may make sense to assume that lot frontage depends on the neighborhood it's in
#Let's try with domain knowledge: if I have a Lot Frontage missing, I can fill in for the value of
#that neighborhood (mean or median)
```

```python
df.groupby('Neighborhood')['Lot Frontage'].mean() 
#This way I get the average lot frontage for each neighborhood
#If I miss the lot frontage for a certain row, I fill in with the mean for its neighborhood
#This is possible with Pandas transform function
```

```python
df['Lot Frontage'] = df.groupby('Neighborhood')['Lot Frontage'].transform(lambda value: value.fillna(value.mean()))
#this is as applying the groupby().mean() using the transform on the NaN values of the Lot Front feature
```

```python
df.isnull().sum()
#The 3 null values for Frontage are those for which it didn't exist a mean for their neighborhood
#I assume that for those this feature is not applicable, so I can fill with 0
df['Lot Frontage'] = df['Lot Frontage'].fillna(0)
```
